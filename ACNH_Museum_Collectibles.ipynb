# 1.Extract
    Animal Crossing Datasets (in CSV) from Kaggle. API from ACNH API. Getting the important datapoints for all the museum collectibles.

```python
# Import Dependencies
import pandas as pd
import requests
import json
from sqlalchemy import create_engine
import datetime
```

```python
# Dictionary for all the Fish, Incects, Fossils and Arts
# Data Sourced from Kaggle
df_fish = pd.read_csv("AC_Resources/fish.csv")
df_fossils = pd.read_csv("AC_Resources/fossils.csv")
df_insects = pd.read_csv("AC_Resources/insects.csv")
df_art = pd.read_csv("AC_Resources/art.csv")

# Unified columns for Fish and Insects
df_cols = ['#','Name','Sell','Internal ID']

# Data Sourced from ACNH API
fish_url = "http://acnhapi.com/v1/fish/"
ins_url = "http://acnhapi.com/v1/bugs/"
fossils_url = 'http://acnhapi.com/v1/fossils/'
art_url = 'http://acnhapi.com/v1/art/'
```

# 2.Transform
Transforming the data into three different datasets: 

1.Alive collectibles - Fish & Insects 

2.Art pieces

3.Fossils

```python
# #,Name, Sell, Where/How, Where, Rarity, Weather,
# Lists for the fish_df 
fish = df_fish[df_cols]
fish_month_northern = []
fish_month_southern = []
fish_time = []
fish_isAllDay = []
fish_isAllYear = []
fish_location = []
fish_rarity = []
fish_image_url=[]
```

```python
for tag in fish['#']:
    target_url = fish_url + str(tag)
    response = requests.get(target_url).json()
    fish_month_northern.append(response['availability']['month-northern'])
    fish_month_southern.append(response['availability']['month-southern'])
    fish_time.append(response['availability']['time'])
    fish_isAllDay.append(response['availability']['isAllDay'])
    fish_isAllYear.append(response['availability']['isAllYear'])
    fish_location.append(response['availability']['location'])
    fish_rarity.append(response['availability']['rarity'])
    fish_image_url.append(response['image_uri'])
```

```python
fish['Month Northern'] = fish_month_northern
fish['Month Southern'] = fish_month_southern
fish['Time'] = fish_time
fish['AllDay T/F'] = fish_isAllDay
fish['AllYear T/F'] = fish_isAllYear
fish['Location'] = fish_location
fish['Rarity'] = fish_rarity
fish['Image URL'] = fish_image_url
```

```python
# Show the fish dataset
fish.head()
```

```python
# Lists for the insect_df
insects = df_insects[df_cols]
ins_month_northern = []
ins_month_southern = []
ins_time = []
ins_isAllDay = []
ins_isAllYear = []
ins_location = []
ins_rarity = []
ins_image_url=[]
```

```python
for tag in insects['#']:
    target_url = ins_url + str(tag)
    response = requests.get(target_url).json()
    ins_month_northern.append(response['availability']['month-northern'])
    ins_month_southern.append(response['availability']['month-southern'])
    ins_time.append(response['availability']['time'])
    ins_isAllDay.append(response['availability']['isAllDay'])
    ins_isAllYear.append(response['availability']['isAllYear'])
    ins_location.append(response['availability']['location'])
    ins_rarity.append(response['availability']['rarity'])
    ins_image_url.append(response['image_uri'])
```

```python
insects['Month Northern'] = ins_month_northern
insects['Month Southern'] = ins_month_southern
insects['Time'] = ins_time
insects['AllDay T/F'] = ins_isAllDay
insects['AllYear T/F'] = ins_isAllYear
insects['Location'] = ins_location
insects['Rarity'] = ins_rarity
insects['Image URL'] = ins_image_url
```

```python
# Show the insects Dataset
insects
```

```python
# Combined both Fish and insects data
live_df = fish.append(insects)
live_df
```

```python
fossils = df_fossils[['Name','Buy','Sell','Source','Internal ID']]
fossils_icon = []
fossils['Name'] = fossils['Name'].replace(' ', '_', regex=True)
fossils
```

```python
for name in fossils['Name']:
    try:
        target_url = fossils_url + name
        response = requests.get(target_url).json()
        fossils_icon.append(response['image_uri'])
    except json.decoder.JSONDecodeError:
        pass
        
```

```python
fossils['Image URL']=pd.Series(fossils_icon)
```

```python
# Show the fossils dataset
fossils.head()
```

```python
genuine_art = df_art.loc[df_art['Genuine']=='Yes']
art = genuine_art[['Name','Category','Real Artwork Title','Artist','Tag','Internal ID']]
art.reset_index(drop=True, inplace=True)
art.index += 1
art.index.name='ID'
art_uri = []
```

```python
for name in list(art.index):
    target_url = art_url + str(name)
    response = requests.get(target_url).json()

    art_uri.append(response['image_uri'])
    
```

```python
art['Image URL'] = art_uri
```

```python

live_df.to_csv('Museum_Output/Alive_Collectibles.csv')

fossils.to_csv('Museum_Output/Fossils_Collectibles.csv')

art.to_csv('Museum_Output/Art_Collectibles.csv')
```

```python
# Establish the connection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection

db = client.museum_collectibles
alive_collection = db.alive_collectibles
fossil_collection = db.fossil_collectibles
art_collection = db.art_collectibles
```

```python
# Insert Into MongoDB
alive_collectibles_data = live_df.to_dict(orient='records')
print(f"Posting {len(alive_collectibles_data)} documents into collection of bestsellers inside \
alive_collectibles_data mongo database...")
fossil_data = fossils.to_dict(orient='records')
print(f"Posting {len(fossil_data)} documents into collection of bestsellers inside \
alive_collectibles_data mongo database...")
art_data = art.to_dict(orient='records')
print(f"Posting {len(art_data)} documents into collection of bestsellers inside \
alive_collectibles_data mongo database...")


alive_collection.insert_many(alive_collectibles_data)
fossil_collection.insert_many(fossil_data)
art_collection.insert_many(art_data)









```
