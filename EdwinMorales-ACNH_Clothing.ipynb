```python
#import dependecies
import pymongo
from pymongo import MongoClient
```

```python
#create mongo db for bags database
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#check if the database exists already
dblist = client.list_database_names()
if "final_bags_df" in dblist:
    client.drop_database("final_bags_df")
```

```python
#define database & the collection
db = client.final_bags_df
collection = db.bags_df
```

```python
mongo_bags = pd.read_csv('final_bag_df.csv')
mongo_bags.head()
```

```python
#create list of dictionaries that hold mongodb docs to be inserted
post = mongo_bags.to_dict(orient='records')
print(f"displaying {len(post)} documents into collection of availability inside mongo database")

#insert list of docs into the db
collection.insert_many(post)
#display results
print(mongo_bags)
```

```python
#verify
results = collection.find()
for result in results:
    print(result)
```
